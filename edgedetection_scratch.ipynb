```python
import numpy as np
import cv2
```

```python
image_path = "/content/dogphoto.png"

grey = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
print(grey)
```

```python
grey
```

```python
rows, cols = grey.shape
```

```python
sobel_vertical = np.array([[-1,0,-1],
                           [-2,0,2],
                           [-1,0,1]])
sobel_vertical
```

```python
k = sobel_vertical.shape[0]
k
```

```python
output_matrix = np.zeros((rows-k+1,cols-k+1))
output_matrix
```

```python
for i in range(1,rows-1):
  for j in range(1, cols-1):
    region = grey[i-1:i+2,j-1:j+2]
    output_matrix[i-1,j-1] = np.sum(region * sobel_vertical)
```

```python
output_matrix
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure = plt.figure(figsize= (10,5))

plt.subplot(1,2,1)
plt.title("original")
plt.imshow(grey, cmap = 'gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("sobel conv edge output")
plt.imshow(output_matrix, cmap= 'gray')
plt.axis('off')

plt.show()
```
